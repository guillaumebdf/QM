```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
vegan_df = pd.read_csv("vegan.csv")
veganism_df = pd.read_csv("veganism.csv")
b12_df = pd.read_csv("b12.csv")
```

```python
vegan = vegan_df.vegan
veganism = veganism_df.veganism
b12 = b12_df.b12
vegan_df.months = pd.to_datetime(vegan_df.months)
veganism_df.months = pd.to_datetime(veganism_df.months)
b12_df.months = pd.to_datetime(b12_df.months)
```

```python
fig = plt.figure(figsize = (10,5))

ax = fig.add_subplot()

ax.plot(vegan_df.months, vegan_df.vegan, 'r', label="search: vegan")
ax.plot(veganism_df.months, veganism_df.veganism, 'b',label="search: veganism")
ax.plot(b12_df.months, b12_df.b12, 'g',label="search: b12")

ax.legend()
ax.set_xlabel("Time")
ax.set_ylabel("Search volume")
ax.set_title("Amount of google searches every day for terms linked to a vegan lifestyle, from 2004 to 2021", size = 14)
```

```python
search = vegan_df.vegan + b12_df.b12 + veganism_df.veganism
```

```python
fig = plt.figure(figsize = (10,5))
ax = fig.add_subplot()

ax.plot(vegan_df.months, search, 'r', label="search: vegan, veganism, b12")

ax.legend()
ax.set_xlabel("Time")
ax.set_ylabel("Search volume")
ax.set_title("Cumulated google searches every day for terms linked to a vegan lifestyle, from 2004 to 2021", size = 14)
```
